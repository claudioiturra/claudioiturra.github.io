__Intro del Curso__

<div style="text-align: justify">El curso ADJM introducirá el uso de la plataforma Jupyter Notebook mediante el análisis de datos oceanográficos. Jupyter Notebook es una gran herramienta para la programación interactiva. La plataforma permite combinar códigos, resultados de simulación, descripciones como comentarios y ecuaciones, manuscritos y presentaciones en un solo archivo de forma ordenada. La plataforma es compatible con variados lenguajes de programación (Matlab, Python, R, etc), el lenguaje que utilizaremos durante el curso sera Matlab. Este curso introductorio se basa en el primer segmento de los cursos Ocean/Atmosphere Time Series Analysis: Theory and Practice y Fundamentals of Ocean/Atmosphere Data Analysis, ambos dictados por The Alfred Wegener Institute.</div>    

<br/>

<div style="text-align: justify">El curso esta orientado para estudiantes de postgrado en Oceanografía (Ciencias del Mar), Ciencias Atmosféricas y Clima. El curso requiere que los participantes tengan nociones básicas del lenguaje de programación Matlab. Utilizando Jupyter Lab, el curso comenzara con una breve descripción de la plataforma mediante una clase interactiva de un tópico de interés (Ej; Series de Furier). Luego de conocer las características de la plataforma mediante variados ejemplos, trabajaremos analizando datos oceanográficos de satélites (temperatura, clorofila y viento), derivadores y perfiladores Argo (asociados al área de interés de los participantes). Los participantes podrán traer sus propias bases de datos, con el fin de generar una breve presentación utilizando la plataforma de trabajo y utilizando los datos analizados. La modalidad del curso será de dos secciones intensivas con una duración de un día de trabajo. Para participar del curso es necesario enviar un correo a claudioiturra@oceanografia.udec.cl, con tu nombre, programa de estudio, breve descripción de tu investigación (*incluir área de estudio*) y tu dirección de correo (Cupos Limitados). Una vez registrado en el curso te haré llegar un correo con información esencial de las actividades a realizar. El curso se dictara de forma presencial en el campus San Andrés de La Universidad Católica de la Santísima Concepción (UCSC).</div>
